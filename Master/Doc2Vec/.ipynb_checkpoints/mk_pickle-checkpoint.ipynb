```python
# モデルのロード
d2v_model = Doc2Vec.load('model/d2v.model')
# d2v_model = Doc2Vec.load('model/test.model')

x=[]
y=[]
z=[]
dic={}
def read_file(path, label):
    print("read_files=", path)
    files=os.listdir(path)
    for file in files:
        if file=='.DS_Store':
            continue
        else:
            z.append(file)                                          # ヘッドライン名
            y.append(label)                                        #ラベル
            x.append(d2v_model.docvecs[file])    # ベクトル
read_file('/Users/Baron/Document/text_mining/Texts/wakati_body/sports-watch', 0)
read_file('/Users/Baron/Document/text_mining/Texts/wakati_body/it-life-hack', 1)
read_file('/Users/Baron/Document/text_mining/Texts/wakati_body/movie-enter', 2)
read_file('/Users/Baron/Document/text_mining/Texts/wakati_body/dokujo-tsushin', 3)


# 保存 --- (*4)
pickle.dump([z, y, x], open('genre.pickle', 'wb'))   # [ ヘッドライン名, カテゴリラベル, ベクトル]
# pickle.dump([z, y, x], open('genre_test.pickle', 'wb'))   # [ ヘッドライン名, カテゴリラベル, ベクトル]
print('ok')
```
